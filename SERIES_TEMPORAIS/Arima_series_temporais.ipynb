```python
# Conversao do formato de datas para dia-mes-ano
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
datas = pd.read_csv('../Dados/Bases_de_Dados/AirPassengers.csv', parse_dates = ['Month'],
                    index_col = 'Month', date_parser = dateparse)
```

```python
datas.head()
```

```python
time_series = datas['#Passengers']
```

```python
time_series # DADOS TRANSFORMADOS EM SERIE TEMPORAL
```

# Decomposicao da Serie temporal

```python
decomposicao = seasonal_decompose(time_series)
tendencias = decomposicao.trend # VERIFICAR A TENDENCIA
sazonal = decomposicao.seasonal # VERIFICAR EFEITO SAZONAL DOS PASSAGEIROS
aleatorio = decomposicao.resid # VERIFICAR SITUACOES INVOLUTARIOS IT IS SAME OUTLIER
```

```python
# GRAFICO DE TENDENCIAS
plt.plot(tendencias) # observa-se crescimento tendencia de crescimento no numero de voos
```

```python
plt.plot(sazonal) # O numero de voos cresce mais na epoca considerada de ferias
```

```python
plt.plot(aleatorio) #Verifica-se num periodo acontece uma situacao involutario ou aleatorio
```

# PREVISOES COM ARIMA

```python
# SUBMISSAO DOS DADOS AO ALGORTIMO * COM BASE NOS PARAMETROS P, Q e D do algoritmo
Treino = auto_arima(time_series, order = (4, 1, 3))
```

```python
previsoes = Treino.predict(n_periods=24) # Previsao dos voos nos proximos 12 meses
previsoes
```

# GRAFICO DAS PREVISOES

```python
# VERIFICAR O TAMANHO
len(time_series)
```

```python
# EXTRAIR OS DADOS DE TREINAMENTO
train = time_series[:130] # do inicio ate 130
train.shape
```

```python
# VISUALIZAR INDICES MINIMO E MAXIMO
train.index.min(), train.index.max()
```

```python
# EXTRAIR DADOS DE TESTE
test = time_series[130:] # de 130 em diante
test.shape
```

```python
# TREINAMENTO
Tr_arima = auto_arima(train, suppress_warnings=True)
```

```python
# PREVISOES EM DATAFRAME
Previsao_arima = pd.DataFrame(Tr_arima.predict(n_periods=14), index=test.index)
Previsao_arima.columns = ['passengers_predictions']
Previsao_arima
```

```python
plt.plot(fig_size = (8, 5))
plt.plot(train, label = 'TRAINING')
plt.plot(test, label = 'TESTING')
plt.plot(Previsao_arima, label = 'PREDICTIONS')
plt.legend();
```
